```python
# Función para analizar ventas por categoría en una tienda
def analyze_category_sales(df, store_name):
    category_sales = df.groupby('Categoría del Producto').size().reset_index(name='Cantidad Vendida')
    print(f"Ventas por Categoría en {store_name}:")
    print(category_sales.sort_values(by='Cantidad Vendida', ascending=False))
    print("-" * 30)

# Analizar ventas por categoría para cada tienda
analyze_category_sales(tienda, 'Tienda 1')
analyze_category_sales(tienda2, 'Tienda 2')
analyze_category_sales(tienda3, 'Tienda 3')
analyze_category_sales(tienda4, 'Tienda 4')
```

```python
import matplotlib.pyplot as plt
def analyze_category_sales_pie(df, store_name):
    category_sales = df.groupby('Categoría del Producto').size().reset_index(name='Cantidad Vendida')

    plt.figure(figsize=(10, 10)) # Aumenta el tamaño de la figura
    plt.pie(category_sales['Cantidad Vendida'], labels=category_sales['Categoría del Producto'], autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribución de Ventas por Categoría en {store_name}', fontsize=16)
    plt.axis('equal')  # Asegura que el gráfico de pastel sea un círculo
    plt.show()
    print("-" * 30)

# Analizar y generar gráficos de pastel de ventas por categoría para cada tienda
analyze_category_sales_pie(tienda, 'Tienda 1')
analyze_category_sales_pie(tienda2, 'Tienda 2')
analyze_category_sales_pie(tienda3, 'Tienda 3')
analyze_category_sales_pie(tienda4, 'Tienda 4')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Combinar los dataframes para facilitar el análisis conjunto
all_stores = pd.concat([
    tienda.assign(Tienda='Tienda 1'),
    tienda2.assign(Tienda='Tienda 2'),
    tienda3.assign(Tienda='Tienda 3'),
    tienda4.assign(Tienda='Tienda 4')
])

# Calcular la calificación promedio por tienda y producto
satisfaccion_cliente = all_stores.groupby(['Tienda', 'Categoría del Producto'])['Calificación'].mean().reset_index()

# Pivotear la tabla para tener las tiendas como columnas
satisfaccion_pivot = satisfaccion_cliente.pivot(index='Categoría del Producto', columns='Tienda', values='Calificación')

# Rellenar los valores NaN con 0 o un valor apropiado si un producto no se vendió en una tienda
satisfaccion_pivot = satisfaccion_pivot.fillna(0) # O usa otro valor como la media general, dependiendo del contexto

# Ordenar las categorías para una mejor visualización si es necesario
satisfaccion_pivot = satisfaccion_pivot.sort_index()

# Crear el gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(15, 8)) # Aumentar el tamaño del gráfico

satisfaccion_pivot.plot(kind='bar', ax=ax, width=0.8)

# Añadir etiquetas y título
ax.set_xlabel('Categoría del Producto')
ax.set_ylabel('Calificación Promedio del Cliente')
ax.set_title('Satisfacción Promedio del Cliente por Categoría de Producto y Tienda')
ax.tick_params(axis='x', rotation=90) # Rotar etiquetas del eje x

# Añadir etiquetas de valor a cada barra (esto puede ser denso, ajustar si es necesario)
# Iterar sobre cada contenedor de barras (uno por tienda)
for container in ax.containers:
    # Iterar sobre cada barra en el contenedor
    for i, bar in enumerate(container):
        # Obtener la altura de la barra (el valor de calificación)
        height = bar.get_height()
        # Si la altura es mayor que 0 (es decir, hay datos para esta categoría/tienda)
        if height > 0:
            # Añadir el texto de la calificación sobre la barra
            ax.text(bar.get_x() + bar.get_width() / 2., # Posición x en el centro de la barra
                    height,                         # Posición y justo encima de la barra
                    f'{height:.2f}',                # Texto a mostrar (calificación formateada)
                    ha='center', va='bottom',        # Alineación horizontal y vertical
                    fontsize=8)                      # Tamaño de fuente

# Ajustar el diseño para evitar que las etiquetas se superpongan
plt.tight_layout()

# Mostrar la leyenda fuera del gráfico si es necesario
ax.legend(title='Tienda', bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
# Identificar los productos más vendidos en cada tienda
def mas_vendidos(df, store_name, top_n=5):
    # Contar la frecuencia de cada producto
    product_counts = df['Producto'].value_counts()

    print(f"\nTop {top_n} Productos más vendidos en {store_name}:")
    print(product_counts.head(top_n))

    # Generar gráfico de barras para el top N
    plt.figure(figsize=(12, 7))
    product_counts.head(top_n).plot(kind='bar', color='skyblue')
    plt.title(f'Top {top_n} Productos más Vendidos en {store_name}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("-" * 30)

# Analizar y graficar los productos más vendidos para cada tienda
mas_vendidos(tienda, 'Tienda 1')
mas_vendidos(tienda2, 'Tienda 2')
mas_vendidos(tienda3, 'Tienda 3')
mas_vendidos(tienda4, 'Tienda 4')
```

```python
import matplotlib.pyplot as plt
def menos_vendidos(df, store_name, bottom_n=5):
    # Contar la frecuencia de cada producto
    product_counts = df['Producto'].value_counts()

    print(f"\nTop {bottom_n} Productos menos vendidos en {store_name}:")
    print(product_counts.tail(bottom_n)) # Usa .tail() para obtener los menos frecuentes

    # Generar gráfico de barras para el bottom N
    plt.figure(figsize=(12, 7))
    # Ordenar por cantidad ascendente para graficar los menos vendidos
    product_counts.tail(bottom_n).sort_values(ascending=True).plot(kind='bar', color='salmon')
    plt.title(f'Top {bottom_n} Productos Menos Vendidos en {store_name}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("-" * 30)

# Analizar y graficar los productos menos vendidos para cada tienda
menos_vendidos(tienda, 'Tienda 1')
menos_vendidos(tienda2, 'Tienda 2')
menos_vendidos(tienda3, 'Tienda 3')
menos_vendidos(tienda4, 'Tienda 4')
```

# ¡Extra! Análisis del desempeño geográfico

```python
# se usa IA para generar el mapa

import pandas as pd
import folium
from folium.plugins import HeatMap

# Combina todos los dataframes en uno solo para facilitar el mapeo
# Asegúrate de que ya tienes 'all_stores' creado como en tu código anterior:
# all_stores = pd.concat([...])

# Verifica que las columnas 'lat' y 'lon' existan y no tengan valores nulos
all_stores = all_stores.dropna(subset=['lat', 'lon', 'Precio'])

# Calcula el centro aproximado del mapa (media de lat y lon)
center_lat = all_stores['lat'].mean()
center_lon = all_stores['lon'].mean()

# Crea un mapa base usando Folium
m = folium.Map(location=[center_lat, center_lon], zoom_start=10) # Ajusta el zoom según la distribución

# Agrega marcadores para cada tienda (opcional, si quieres ver la ubicación exacta de las tiendas)
# Puedes añadir las coordenadas manualmente si las conoces, o usar las medias de lat/lon si son representativas
# Ejemplo con las medias:
store_locations = all_stores.groupby('Tienda')[['lat', 'lon']].mean().reset_index()

for index, row in store_locations.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=row['Tienda'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)


# Crea una capa de mapa de calor para visualizar la densidad de ventas (usando 'Precio' como peso)
# El formato para HeatMap es [[lat, lon, weight], [lat, lon, weight], ...]
heat_data = [[row['lat'], row['lon'], row['Precio']] for index, row in all_stores.iterrows()]

HeatMap(heat_data).add_to(m)

# Muestra el mapa
m
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Calcula la facturación total por tienda
facturacion_por_tienda = all_stores.groupby('Tienda')['Precio'].sum().reset_index()

# Obtén las coordenadas promedio para cada tienda
store_locations = all_stores.groupby('Tienda')[['lat', 'lon']].mean().reset_index()

# Combina la facturación con las ubicaciones
sales_by_location = pd.merge(facturacion_por_tienda, store_locations, on='Tienda')

print("\nDatos para el gráfico de dispersión (Facturación por Ubicación de Tienda):")
print(sales_by_location)

# Crea el gráfico de dispersión
plt.figure(figsize=(10, 8))

scatter = plt.scatter(sales_by_location['lon'], sales_by_location['lat'],
                      c=sales_by_location['Precio'], # Color basado en la facturación
                      cmap='YlOrRd', # Mapa de colores (amarillo a rojo)
                      s=sales_by_location['Precio']/1000, # Tamaño basado en la facturación (ajusta el divisor si los puntos son muy grandes o pequeños)
                      alpha=0.7, # Transparencia
                      edgecolors='w', # Borde blanco para mayor visibilidad
                      linewidths=0.5)

# Añadir etiquetas de las tiendas a los puntos
for i, row in sales_by_location.iterrows():
    plt.text(row['lon'], row['lat'], row['Tienda'], fontsize=9, ha='right')


plt.colorbar(scatter, label='Facturación Total')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Concentración de Facturación por Ubicación de Tienda')
plt.grid(True)
plt.show()

```
